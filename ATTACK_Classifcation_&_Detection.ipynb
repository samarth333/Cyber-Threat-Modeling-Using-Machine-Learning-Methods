```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.optimizers import RMSprop
from sklearn.preprocessing import StandardScaler
```

```python
# Load the training and testing datasets
train_data = pd.read_csv('big_data.csv')
test_data = pd.read_csv('/content/test (1).csv')
```

```python
# Extract the features and labels from the datasets
train_features = train_data.iloc[:, :-1].values
train_label = train_data.iloc[:, -1].values
test_features = test_data.iloc[:, :-1].values
test_label = test_data.iloc[:, -1].values
```

```python
# Convert label to one-hot encoded vectors
num_classes = len(np.unique(train_label))
train_label = np.eye(num_classes)[train_label]
test_label = np.eye(num_classes)[test_label]

# Define the RNN model
model = Sequential()
model.add(LSTM(64, input_shape=(train_features.shape[1], 1)))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])

# Train the model
model.fit(train_features.reshape(-1, train_features.shape[1], 1), train_label, batch_size=64, epochs=100, verbose=0)

# Evaluate model on testing dataset and compute evaluation metrics
test_pred = model.predict(test_features.reshape(-1, test_features.shape[1], 1))
test_pred_class = np.argmax(test_pred, axis=1)
test_true_class = np.argmax(test_label, axis=1)

# Print predicted and true output for test data
print('\nPredicted output for test data:')
for i in range(len(test_true_class)):
    print(f'True: {class_names[test_true_class[i]]}, Predicted: {class_names[test_pred_class[i]]}')

# Compute confusion matrix and print evaluation metrics
confusion_mat = confusion_matrix(test_true_class, test_pred_class)
print('\nConfusion matrix:')
print(confusion_mat)
class_names = ['class_0', 'class_1', 'class_2']
print('\nClassification report:')
print(classification_report(np.argmax(test_label, axis=1), np.argmax(test_pred, axis=1), target_names=class_names))


accuracy = np.sum(test_true_class == test_pred_class) / len(test_true_class)
precision = np.diagonal(confusion_mat) / np.sum(confusion_mat, axis=0)
recall = np.diagonal(confusion_mat) / np.sum(confusion_mat, axis=1)
f1_score = 2 * precision * recall / (precision + recall)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {np.mean(precision):.4f}')
print(f'Recall: {np.mean(recall):.4f}')
print(f'F1 score: {np.mean(f1_score):.4f}')
```
